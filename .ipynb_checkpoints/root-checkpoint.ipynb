```python
import pandas as pd
import yfinance as yf
import numpy as np
```

```python
x = yf.download(tickers = "^NSEI",  # list of tickers
            period = "20y",         # time period
            interval = "1d",       # trading interval
            prepost = False,       # download pre/post market hours data?
            repair = True)
```

```python
x.tail()
```

```python
x = pd.DataFrame(x)
```

```python
x = x.drop(["Adj Close","Volume","Repaired?"]   , axis = 1)
```

```python
x["H20"] = x['High'].rolling(window=20).max()
x["H20"] = x["H20"].shift(1)
```

```python
x["L20"] = x['Low'].rolling(window=20).min()
x["L20"] = x["L20"].shift(1)
```

```python
x["H10"] = x['High'].rolling(window=10).max()
x["H10"] = x["H10"].shift(1)
```

```python
x["L10"] = x['Low'].rolling(window=10).min()
x["L10"] = x["L10"].shift(1)
```

```python
x["H55"] = x['High'].rolling(window=55).max()
x["H55"] = x["H55"].shift(1)
```

```python
x['D1'] = abs(x['High'] - x['Close'].shift())
x['D2'] = abs(x['Low']-x['Close'].shift())
x['D3'] = x['High'] - x['Low']
x['TR'] = x[["D1", "D2" , "D3"]].max(axis=1)

x = x.drop(["D1", "D2" , "D3"]   , axis = 1)

x['ATR'] = np.nan
```

```python

range_atr = 20
```

```python
x['ATR'][range_atr] = x.iloc[1:(range_atr+1)]['TR'].mean()
for i in range (range_atr+1 , x.shape[0]):
    x['ATR'][i] =  ((x['ATR'][i-1]*(range_atr-1))+x['TR'][i])/range_atr
```
